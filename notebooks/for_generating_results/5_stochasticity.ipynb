# Spasmodic recruitment stochasticity

```python
from rl4fisheries.envs.asm_fns import get_r_devs_v2, get_r_devs
```

```python
import matplotlib as mpl
import pandas as pd
import seaborn as sns
```

```python
N_YEAR = 1_000

r_v1 = pd.DataFrame({
    'r deviation': get_r_devs(n_year=N_YEAR),
    't': list(range(N_YEAR)),
})
r_v1['Stochasticity'] = 'Lognormal + spasms'
#
r_v2 = pd.DataFrame({
    'r deviation': get_r_devs_v2(n_year=N_YEAR),
    't': list(range(N_YEAR)),
})
r_v2['Stochasticity'] = 'Lognormal'


rdevs = pd.concat([r_v1, r_v2], ignore_index=True)
```

```python
sns.set_theme(
    palette=sns.color_palette("husl", 6),
    font="Times New Roman",
    font_scale=1.5,
    style="whitegrid",
)
sns.axes_style(
    style = {"axes.linewidth":4}
)
```

```python
g = sns.relplot(
    data=rdevs,
    x='t',
    y='r deviation',
    row='Stochasticity',
    hue='Stochasticity',
    kind='line',
    aspect=2.,
    linewidth=2.8,
    # facet_kws={'margin_titles': True}
)

sns.move_legend(
    g, "lower center",
    bbox_to_anchor=(0.41, -0.05), frameon=False, ncol=2, title=None,
)

mpl.rcParams['axes.linewidth'] = 4
for ax in g.axes.ravel():
    ax.set_yscale('log')

    ax.spines['right'].set_visible(True)
    ax.spines['left'].set_visible(True)
    ax.spines['top'].set_visible(True)
    ax.spines['bottom'].set_visible(True) 
```

```python
r_v1.plot(logy=True)
```
